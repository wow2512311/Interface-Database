```python
import numpy as np
import matplotlib.pyplot as plt

directory = "./ProjectedDensityOfStates/zigzag/Z-4-3-7-1-DOS.npy"
data = np.load(directory,allow_pickle=True).item()
fig = plt.figure(figsize=(25,10))
dos = data['dos']
energies = data['energies']
Projected_DensityOfStates_GridValues = data['Projected_DensityOfStates_GridValues']
plt.title('Z-4-3-7-1-DOS',fontproperties='Times New Roman', size=40)
plt.xlabel(r'Energy (eV)',fontproperties='Times New Roman', size=30)
plt.ylabel(r'DOS (eV$^{-1}$)',fontproperties='Times New Roman', size=30)

plt.xticks([energies[0],energies[200],energies[400],energies[600],energies[800],energies[1000]],fontproperties='Times New Roman', size=25)
plt.yticks(fontproperties='Times New Roman', size=25)
plt.grid()
plt.xlim(-6,6)
plt.ylim(0,600)
x = energies
y = dos
plt.plot(x,y,color='k',label='Total')
y1 = Projected_DensityOfStates_GridValues[0,:]
y2 = Projected_DensityOfStates_GridValues[1,:]
plt.plot(x,y1,label='Left')
plt.plot(x,y2,label='Right')
legend_font = {
     'family': 'Times New Roman',  
     'style': 'normal',
     'size': 
     'weight': "normal",
 }
plt.legend(fontsize="20",prop = legend_font)
```
